```python
import pandas as pd

file_path = 'AmazonSalesData.csv'
data = pd.read_csv(file_path)
```

```python
# inspect data

print(display(data.head()))
print(data.dtypes)
print(data.describe())
```

```python
print(data.isnull().sum())
print(data.duplicated().sum())
```

```python
data = data.rename(columns={'Order Date': 'order_date', 
                            'Order ID': 'order_id', 
                            'Ship Date': 'ship_date',
                            'Country': 'country', 
                            'Item Type': 'item_type',
                            'Total Profit': 'total_profit',
                            'Total Cost': 'total_cost',
                            'Total Revenue': 'total_revenue',
                            'Unit Price': 'unit_price',
                            })

print(display(data.head()))
```

```python
# cleaned data
cleaned_file_path = 'Cleaned_AmazonSalesData.csv'
data.to_csv(cleaned_file_path, index=False)
```
